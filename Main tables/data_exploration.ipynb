```python
import pandas as pd 
listings_df = pd.read_csv("listings.csv")
listings_df
```

```python
def missing_values_table(df):
    mis_val = df.isnull().sum()
    mis_val_percent = 100 * df.isnull().sum() / len(df)
    mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
    mis_val_table_ren_columns = mis_val_table.rename(
    columns = {0 : 'Missing Values', 1 : '% of Total Values'})
    mis_val_table_ren_columns = mis_val_table_ren_columns[
        mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(
    '% of Total Values', ascending=False).round(1)
    print ("Your selected dataframe has " + str(df.shape[1]) + " columns.\n"      
        "There are " + str(mis_val_table_ren_columns.shape[0]) +
            " columns that have missing values.")
    return mis_val_table_ren_columns
```

```python
missing_values_table(listings_df)
```

```python
missing_values_table(listings_df.dropna())
```

```python
listings_df.dropna().to_csv("clean_listings.csv")
```

```python
calendar_df = pd.read_csv("calendar.csv")
calendar_df
```

```python
calendar_df = calendar_df.dropna()
calendar_df["price"] = calendar_df["price"].apply(lambda current_number: int(current_number))
calendar_df["adjusted_price"] = calendar_df["adjusted_price"].apply(lambda current_number: int(current_number))
calendar_df["minimum_nights"] = calendar_df["minimum_nights"].apply(lambda current_number: int(current_number))
calendar_df["maximum_nights"] = calendar_df["maximum_nights"].apply(lambda current_number: int(current_number))

calendar_df.to_csv("clean_calendar.csv", index = False)
```
